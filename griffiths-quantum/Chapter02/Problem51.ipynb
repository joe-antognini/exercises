```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-5, 5, num=100)
V = -1 / np.cosh(X)**2
```

```python
plt.plot(X, V, c='k')
plt.xlabel('$ax$')
plt.ylabel('$V$');
```

```python
plt.plot(X, 1 / np.sqrt(2) / np.cosh(X), c='k')
plt.xlabel('$ax$')
plt.ylabel(r'$\psi_0$');
```
