```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
```

```python
def simulate_and_plot_flips(p, N=1000):
    flips = np.random.binomial(1, p, size=N)
    Ns = np.arange(N) + 1
    proportions = np.cumsum(flips) / Ns
    
    plt.plot(Ns, proportions)
    plt.axhline(p, linestyle=':', color='gray')
    plt.xlabel(r'$N$')
    plt.ylabel('Proportion');
```

```python
simulate_and_plot_flips(p=0.3)
```

```python
simulate_and_plot_flips(p=0.03)
```
