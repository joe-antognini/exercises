## Part b.

In the non-interacting limit the trajectory is on a torus.  If it were a sphere it would be possible to set the (normalized) coordinates of all dimensions to 0 except for 1.  However, the x or y position of Earth must be non-zero as well as that of Jupiter.  So it cannot be a sphere.

Imagine the x-y axis of Jupiter, and then put the x-coordinate of Earth on the z-axis.  Then we see that the trajectory will wind around the x-y plane and oscillate in the z-axis.

The Earth would wind around this surface 11.7 times every Jovian year.

### Part c.

```python
masses = np.logspace(-1, 0, num=100)
```

```python
def is_unstable(m_jupiter, t_integrate=10_000):
    sim = rebound.Simulation()
    sim.add(m=1)
    sim.add(m=3e-6, a=1, e=0.01671)
    sim.add(m=9.55e-4, a=5.2034, e=0.04839)
    sim.particles[2].m = m_jupiter
    sim.integrate(t_integrate)
    
    return sim.orbits()[0].e > 1
```

```python
import multiprocessing
```

```python
with multiprocessing.Pool() as pool:
    stability_results = list(pool.map(is_unstable, masses))
```

```python
masses[np.argmax(stability_results)]
```

## Part d.

```python
sim = rebound.Simulation()
sim.add(m=1)
sim.add(m=3e-6, a=1, e=0.01671)
sim.add(m=9.55e-4, a=5.2034, e=0.04839)
sim.move_to_com()

poincare_sections = []

prev_angle = 0
prev_positions = np.array([[p.x, p.y] for p in sim.particles])
prev_positions -= prev_positions[0]
prev_positions = prev_positions[1:]
prev_vel = np.array([[p.vx, p.vy] for p in sim.particles])
prev_vel -= prev_vel[0]
prev_vel = prev_vel[1:]

for _ in tqdm(range(10_000)):
    sim.step()
    cur_positions = np.array([[p.x, p.y] for p in sim.particles])
    cur_positions -= cur_positions[0]
    cur_positions = cur_positions[1:]
    cur_vel = np.array([[p.vx, p.vy] for p in sim.particles])
    cur_vel -= cur_vel[0]
    cur_vel = cur_vel[1:]
    
    cur_angles = np.arctan2(
        cur_positions[:, 1],
        cur_positions[:, 0],
    )
    cur_angle = cur_angles[1] - cur_angles[0]
    
    if cur_angle > 0 and prev_angle < 0:
        f = cur_angle / (cur_angle - prev_angle)
        mean_pos = (1 - f) * cur_positions + f * prev_positions
        mean_vel = (1 - f) * cur_vel + f * prev_vel
        poincare_r = np.linalg.norm(mean_pos[1] - mean_pos[0])
        poincare_v = np.dot(mean_vel[0], mean_pos[1] - mean_pos[0]) / poincare_r 
        poincare_pos = ((1 - f) * cur_positions[0] + f * cur_positions[1])
        poincare_sections.append([poincare_r, poincare_v])
    
    prev_angle = cur_angle
    prev_positions = cur_positions[:]
    prev_vel = cur_vel[:]

poincare_sections = np.array(poincare_sections)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(poincare_sections[:, 0], poincare_sections[:, 1], s=10, edgecolor='none')
plt.xlabel('Distance from Earth to Jupiter')
plt.ylabel('Velocity of Earth toward Jupiter')
plt.title('Poincare section');
```
