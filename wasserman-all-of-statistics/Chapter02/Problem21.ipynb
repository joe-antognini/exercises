```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
def max_y(n: int, beta: float = 1, n_samples: int = 10_000):
    x_uniform = np.random.uniform(0, 1, size=(n, n_samples))
    sample = scipy.stats.expon.ppf(x_uniform)
    return np.max(sample, axis=0)
```

```python
def max_y_pdf(x, n, beta):
    return n / beta * np.exp(-x / beta) * (1 - np.exp(-x / beta))**(n - 1)
```

```python
plt.figure(figsize=(8, 8))
X = np.linspace(0, 10, num=1000)

for i, n in enumerate(range(1, 18, 2)):
    plt.subplot(3, 3, i+1)
    
    max_ys = max_y(n, n_samples=100_000)
    
    plt.hist(max_ys, 100, histtype='step', density=True)
    plt.plot(X, max_y_pdf(X, n, beta=1))
    plt.xlim([0, 10])
    plt.title(f'n = {n}')
plt.tight_layout();
```
