```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.uniform(0, 1, size=1_000_000)
Y = np.random.uniform(0, 1, size=1_000_000)
```

```python
x1 = np.linspace(-1, 0, num=100)
x2 = np.linspace(0, 1, num=100)

plt.hist(X - Y, 1000, density=True)
plt.plot(x1, 1 + x1, color='k')
plt.plot(x2, 1 - x2, color='k');
```

```python
x1 = np.linspace(0, 1, num=1000)
x2 = np.linspace(1, 100, num=1000)

plt.hist(X/Y, bins=np.linspace(0, 100, num=1000), density=True)
plt.plot(x1, np.full_like(x1, 1/2), color='k')
plt.plot(x2, 1 / (2 * x2**2), color='k')
plt.xlim([0, 10]);
```
