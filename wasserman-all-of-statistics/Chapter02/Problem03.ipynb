# Problem 2.3

Let $A = (-\infty, x]$ and let $B_i = (-\infty, y_i)$ where

$$
y_1 < y_2 < \cdots < x
$$

and

$$
\lim_{i\to\infty} y_i = x
$$

Now $\mathbb{P}(\omega \in A) = \mathbb{P}(X \leq x) = F(x)$ and $\mathbb{P}(\omega \in B_i) = \mathbb{P}(X < y_) = F(y_i)$.  Consider the set $A - B_i$.  By definition this is disjoint with $B_i$, so

$$
\mathbb{P}(\omega \in A - B_i) = \mathbb{P}(y_i \leq X \leq x)
$$

and

$$
\mathbb{P}(\omega \in A) = \mathbb{P}(\omega \in A - B_i) + \mathbb{P}(\omega \in B_i)
$$

$$
\begin{eqnarray}
\implies \mathbb{P}(\omega \in A - B_i) & = & \mathbb{P}(\omega \in A) - \mathbb{P}(\omega \in B_i) \\
& = & F(x) - F(y_i)
\end{eqnarray}
$$

Now $\lim_{i\to\infty} A - B_i = {x}$, so

$$
\begin{eqnarray}
\lim_{i\to\infty} \mathbb{P}(\omega \in A - B_i) = \mathbb{P}(X = x) = \lim_{i\to\infty} F(x) - F(y_i) \\
& = & F(x) - F(x^-)
\end{eqnarray}
$$

Next we wish to show that $\mathbb{P}(x < X \leq y) = F(y) - F(x)$.

Let $A = (-\infty, x]$ and $B = (-\infty, y]$.

$$
\begin{eqnarray}
\mathbb{P}(\omega\in A) & = & \mathbb{P}(X \leq x) & = & F(x) \\
\mathbb{P}(\omega\in B) & = & \mathbb{P}(X \leq y) & = & F(y) \\
\end{eqnarray}
$$

Now consider $B - A$.  By definitio nthis forms a disjoint partition with $A$, so

$$
\mathbb{P}(\omega \in B) = \mathbb{P}(\omega \in A) + \mathbb{P}(\omega \in B - A)
$$

But $\mathbb{P}(\omega \in B - A) = \mathbb{P}(x < X \leq y)$, so

$$
\begin{eqnarray}
\mathbb{P}(x < X \leq y) & = & \mathbb{P}(X \leq y) - \mathbb{P}(X \leq x) \\
& = & F(y) - F(x)
\end{eqnarray}
$$

----

Now we wish to show that $\mathbb{P}(X > x) = 1 - F(x)$.  Let $A = (-\infty, x]$.  Now $\mathbb{P}(\omega \in A) = 1 - \mathbb{P}(\omega \in A^c)$, but $A^c = (x, \infty)$.  So,

$$
\begin{eqnarray}
\mathbb{P}(X > x) & = & 1 - \mathbb{P}(X \leq x) \\
& = & 1 - F(x)
\end{eqnarray}
$$

Lastly, we wish to show that if $X$ is continuous,

$$
\begin{eqnarray}
F(b) - F(a) & = & \mathbb{P}(a < X < b) & = & \mathbb{P}(a \leq X < b) \\
& = & \mathbb{P}(a < X \leq b) & = & \mathbb{P}(a \leq X \leq b)
\end{eqnarray}
$$

Now, if $X$ is continuous, then there exists a function $f_X$ such that

$$
\mathbb{P}(a < X < b) = \int_a^b f_X(x)\, dx
$$

Now,

$$
\begin{eqnarray}
\mathbb{P}((a < X \leq b) & = & \mathbb{P}(a < X < b) + \mathbb{P}(X = b) \\
& = & \mathbb{P}(a < X < b) + \lim_{\epsilon \to 0} (b - \epsilon < X < b + \epsilon) \\
& = & \mathbb{P}(a < X < b) + \lim_{\epsilon \to 0} \int_{b-\epsilon}^{b + \epsilon} f_X(x) \, dx \\
& = & \mathbb{P}(a < X < b)
\end{eqnarray}
$$

A similar argument switching $a$ and $b$ shows $\mathbb{P}(a \leq X < b) = \mathbb{P}(a < X < b)$.  Then since $\mathbb{P}(X = a) = \mathbb{P}(X = b) = 0$, we also have $\mathbb{P}(a \leq X \leq b) = \mathbb{P}(a < X < b)$.

Lastly,

$$
\begin{eqnarray}
\mathbb{P}(a < X < b) & = & \int_a^b f_X(x) \, dx \\
& = & \int_{-\infty}^b f_X(x) \, dx - \int_{-\infty}^a f_X(x) \, dx \\
& = & F(b) - F(a)
\end{eqnarray}
$$
