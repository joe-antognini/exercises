## Part a.

$$
\begin{eqnarray}
\psi_2 & = & \frac{1}{\sqrt{2!}} (a_+)^2 \psi_0 \\
& = & \frac{1}{\sqrt{2}} a_+ \psi_1 \\
& = & \frac{1}{\sqrt{2}} \frac{1}{\sqrt{2 \hbar m \omega}} \left(m \omega x - \hbar \frac{\textrm{d}}{\textrm{d} x} \right) \left( \frac{m \omega}{\pi \hbar} \right)^{\frac{1}{4}} \sqrt{\frac{2m \omega}{\hbar}} x \exp \left[ -\frac{m\omega}{2 \hbar}x^2 \right] \\
& = & \frac{1}{\sqrt{2} \hbar} \left( \frac{m\omega}{\pi\hbar} \right)^{\frac{1}{4}} \left[m\omega x^2 - \hbar + 2x^2 \left(\frac{m\omega}{2}\right) \right] \exp \left[-\frac{m\omega}{2\hbar}x^2 \right] \\
& = & \frac{1}{\sqrt{2}} \left(\frac{m\omega}{\pi \hbar}\right)^{\frac{1}{4}} \left[ 2 \left( \frac{m\omega}{\hbar} \right) x^2 - 1 \right] \exp \left[-\frac{m\omega}{2\hbar} x^2 \right]
\end{eqnarray}
$$

## Part b.

```python
import matplotlib.pyplot as plt
import numpy as np
```

We'll plot the functions $\psi$ in units where $m \omega / \hbar = 1$.

```python
X = np.linspace(-5, 5, num=1000)
psi0 = 1 / np.pi**(1/4) * np.exp(-X**2 / 2)
psi1 = 1 / np.pi**(1/4) * np.sqrt(2) * X * np.exp(-X**2 / 2)
psi2 = 1 / np.sqrt(2) / np.pi**(1/4) * (2 * X**2 - 1) * np.exp(-X**2 / 2)
```

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.plot(X, psi0, c='k')
plt.xlabel('$x$')
plt.ylabel('$\psi_0$')
plt.axhline(0, color='gray', linestyle=':')

plt.subplot(1, 3, 2)
plt.plot(X, psi1, c='k')
plt.xlabel('$x$')
plt.ylabel('$\psi_1$')
plt.axhline(0, color='gray', linestyle=':')

plt.subplot(1, 3, 3)
plt.plot(X, psi2, c='k')
plt.xlabel('$x$')
plt.ylabel('$\psi_2$')
plt.axhline(0, color='gray', linestyle=':')

plt.tight_layout();
```

## Part c.

First let's consider $\psi_0$ and $\psi_1$.

$$
\begin{eqnarray}
\int_{-\infty}^{\infty} \psi_0 \psi_1 \, dx & = & \int_{-\infty}^{\infty} \left(\frac{m\omega}{\pi \hbar}\right)^{\frac{1}{4}} \exp \left[ -\frac{m\omega}{2\hbar}x^2 \right] \left( \frac{m\omega}{\pi \hbar} \right)^{\frac{1}{4}} \sqrt{\frac{2m\omega}{\hbar}}x \exp \left[ -\frac{m\omega}{2\hbar} x^2 \right] \, \textrm{d}x \\
& \propto & \int_{-\infty}^{\infty} x \exp \left[ -\frac{m\omega}{2\hbar}x^2 \right] \, \textrm{d}x \\
& = & 0
\end{eqnarray}
$$

where the last step is due to the fact that we are taking a symmetric integral of an odd function.

Next we'll consider $\psi_1$ and $\psi_2$:

$$
\begin{eqnarray}
\int_{-\infty}^{\infty} \psi_1 \psi_2 \, \textrm{d}x & \propto & \int_{-\infty}^{\infty} x \exp \left[-\frac{m\omega}{2\hbar} x^2 \right] \left[2 \left( \frac{m\omega}{\hbar} \right) x^2 - 1 \right] \exp \left[-\frac{m\omega}{2\hbar}x^2 \right] \, \textrm{d}x \\
& = & 0 \\
\end{eqnarray}
$$

where again we have a symmetric integral of an odd function.

Finally, we'll consider $\psi_0$ and $\psi_2$.  For simplicity let's define

$$
\alpha \equiv \frac{m\omega}{\hbar}
$$

So we have

$$
\begin{eqnarray}
\int_{-\infty}^{\infty} \psi_0 \psi_2 \, \textrm{d}x & = & \int_{-\infty}^{\infty} \left( \frac{\alpha}{\pi} \right)^{\frac{1}{4}} e^{-\frac{\alpha}{2}x^2} \frac{1}{\sqrt{2}} \left(\frac{\alpha}{\pi} \right)^{\frac{1}{4}} \left(2 \alpha x^2 - 1 \right) e^{-\frac{\alpha}{2} x^2} \, \textrm{d}x \\
& = & \sqrt{\frac{\alpha}{\pi}} \int_{-\infty}^{\infty} \left(2 \alpha x^2 - 1 \right) e^{-\alpha x^2} \, \textrm{d}x \\
& = & \sqrt{\frac{\alpha}{2\pi}} \left( 2 \alpha \frac{\sqrt{\pi}}{4 \alpha^{\frac{3}{2}}} - \frac{1}{2} \sqrt{\frac{\pi}{\alpha}} \right) \\
& = & 0
\end{eqnarray}
$$
