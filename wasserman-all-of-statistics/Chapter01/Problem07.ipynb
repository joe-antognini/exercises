# Problem 1.7

We wish to show that

$$
\mathbb{P}\left( \bigcup_{n=1}^{\infty} A_n \right) \leq \sum_{n=1}^{\infty} \mathbb{P}(A_n)
$$

Let

$$
B_n \equiv A_n - \bigcup_{i=1}^{n-1} A_i
$$

Now I claim that the $B_n$ are disjoint.  Consider $B_n$ and $B_m$ and take $n > m$ without loss of generality.  By definition,

$$
B_n \cap \left( \bigcup_{i=1}^{n-1} A_i \right) = \emptyset
$$

But $B_m \subset \bigcup_{i=1}^{n-1} A_i$ since $B_m \equiv A_m - \bigcup_{i=1}^{m-i} A_i$ and $m < n$.  So $A_m \subset \bigcup_{i=1}^{n-1} A_i$.

In generaly, if $X \cap Y = \emptyset$ and $Z \subset Y$, then $X \cap Z = \emptyset$.  So,

$$
B_n \cap B_m = \emptyset
$$

and the $B_n$ are all disjoint.

Moreover, we must also show that

$$
\bigcup_{n=1}^{\infty} A_n = \bigcup_{n=1}^{\infty} B_n
$$

Let us prove this by induction.  For the base case of $n=1$, we simply have $A_1 = B_1$.  Now suppose it holds that

$$
\bigcup_{i=1}^{n-1} A_i = \bigcup_{i=1}^{n-1} B_i
$$

for some $n$.  Consider the $n^{\textrm{th}}$ case.

$$
\begin{eqnarray}
\bigcup_{i=1}^n B_i & = & \left( \bigcup_{i=1}^{n-1} B_i \right) \cup B_n \\
& = & \left( \bigcup_{i=1}^{n-1} A_i \right) \cup \left( A_n - \bigcup_{i=1}^{n-1} A_i \right)
\end{eqnarray}
$$

In generaly, $X \cup (Y - X) = X \cup Y$, so

$$
\begin{eqnarray}
\bigcup_{i=1}^n B_i & = & \left( \bigcup_{i=1}^{n-1} A_i \right) \cup A_n \\
& = & \bigcup_{i=1}^n A_i
\end{eqnarray}
$$

This means that

$$
\mathbb{P}\left( \bigcup_{n=1}^{\infty} A_n \right) = \mathbb{P} \left( \bigcup_{n=1}^{\infty} B_n \right)
$$

$$
\implies \mathbb{P} \left( \bigcup_{n=1}^{\infty} A_n \right) = \sum_{i=1}^{\infty} \mathbb{P}(B_i)
$$

Now, $B_n \subset A_n$, so $\mathbb{P}(B_n) \leq \mathbb{P}(A_n)$.  Hence

$$
\mathbb{P} \left( \bigcup_{n=1}^{\infty} A_n \right) \leq \sum_{n=1}^{\infty} \mathbb{P} (A_n)
$$
