```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 10, num=1000)
x0 = 10
```

```python
x_slow = x0 / np.sqrt(1 + 2 * t * x0**2)
```

```python
x_fast = x0 * np.exp(-t)
```

```python
plt.plot(t, x_slow, label=r'$\dot{x} = -x^3$')
plt.plot(t, x_fast, label=r'$\dot{x} = -x$')
plt.axhline(0, linestyle=':', color='gray')
plt.xlabel('t')
plt.ylabel('x')
plt.legend();
```
