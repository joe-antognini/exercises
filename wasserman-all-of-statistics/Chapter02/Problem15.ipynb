```python
import matplotlib.pyplot as plt
import numpy as np
```

The PDF of the exponential distirbution is

$$
f(x) = \frac{e^{-x/\beta}}{\beta}
$$

The CDF of the exponential distribution is then

$$
\begin{eqnarray}
F(x) & = & \frac{1}{\beta} \int_{0}^x e^{-x/\beta} \, dx \\
& = & 1 - e^{-x/\beta}
\end{eqnarray}
$$

The inverse is then

$$
F^{-1}(u) = -\beta \log (1 - u)
$$

```python
def exp_variates(u, beta=1):
    return -beta * np.log(1-u)
```

```python
U = np.random.uniform(0, 1, size=10_000)
```

```python
exps = exp_variates(U)
```

```python
plt.hist(exps, bins=100, histtype='step', density=True, label='Histogram')

X = np.linspace(0, 10, num=1000)
Y = np.exp(-X)

plt.plot(X, Y, label='PDF')
plt.xlabel('$x$')
plt.ylabel('Probability density')
plt.legend();
```
