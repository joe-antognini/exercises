## Part a.

In order to find the PDF $f_Y$ we must first find the CDF.  Note that $e^X$ is monotonic, so the set of $y$ to be integrated over is simply

$$
\begin{eqnarray}
A_y & = & \{x: e^X \leq y\} \\
& = & \{x : x \leq \log y\}
\end{eqnarray}
$$

Hence

$$
\begin{eqnarray}
F_Y(y) & = & \textbf{P}(e^X \leq y) \\
& = & \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\log y} e^{-x^2/2} \, dx
\end{eqnarray}
$$

The PDF is then

$$
\begin{eqnarray}
f_Y(y) & = & \frac{dF_Y}{dy} \\
& = & \frac{dF_Y}{d\log y} \frac{d \log y}{dy} \\
& = & \frac{e^{-\frac{1}{2}\log^2 y}}{\sqrt{2\pi} y}
\end{eqnarray}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.linspace(0.001, 10, num=1000)
```

```python
def y_pdf(y):
    return np.exp(-np.log(y)**2 / 2) / (np.sqrt(2 * np.pi) * y)
```

```python
plt.plot(y, y_pdf(y))
plt.xlabel('$y$')
plt.ylabel('$f_Y(y)$');
```
